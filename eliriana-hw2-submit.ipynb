```python
#a

import json 
from urllib.request import urlopen

url = 'http://worldcup.sfg.io/matches'
data = urlopen(url).read().decode('utf-8')
eb = json.loads(data)
num_matches = len(eb)

print('Number of matches: ' + str(num_matches))
```

```python
#b 

for i in range(0, len(eb)):
    ebi = eb[i]

    home_country = ebi['home_team']['country']
    home_goals = ebi['home_team']['goals']

    away_country = ebi['away_team']['country']
    away_goals = ebi['away_team']['goals']

    print(home_country + ' - ' + away_country + ' ' + str(home_goals) 
          + ':' + str(away_goals))
```

```python
#c

import pandas as pd
from datetime import datetime

# load JSON into dataframe
dict = eb
keys = ['fifa_id', 'location', 'datetime', 'home_team_country', 'away_team_country', 'winner', 'home_team_statistics', 'away_team_statistics']
fifa = [[match[i] for i in keys] for match in dict] 
fifa_table = pd.DataFrame(fifa)
fifa_table.columns = keys

# reformat date and time
dates = pd.DatetimeIndex(fifa_table['datetime']).date
hours = pd.DatetimeIndex(fifa_table['datetime']).hour
minutes = pd.DatetimeIndex(fifa_table['datetime']).minute
time = [None] * 52
for i in range (0, 51):
    # use zfill to make sure time is always in (hh,mm) format
    time[i] = "(" + str(hours[i]).zfill(2) + ", " + str(minutes[i]).zfill(2) + ")" 

#reformat date and time in dataframe
del fifa_table['datetime']
fifa_table.insert(2, "Date", dates) 
fifa_table.insert(3, "Time", time) 
fifa_table

```

```python
#d 

%matplotlib inline
import matplotlib.pyplot as plt
game_df = fifa_table['location'].value_counts().reset_index()
game_df.columns = ['location', 'counts']
game_df.plot(kind='pie', y =  'counts',  autopct='%d', 
 startangle=90, shadow=False, labels=game_df['location'], legend = False, fontsize=14)
plt.title('Games Played in each Stadium', fontsize=18, weight='bold')
plt.xlabel('');
plt.ylabel('');
```

```python
#e

import seaborn as sns
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize

#get nested json statistic data 
home_data = json_normalize(data= fifa_table['home_team_statistics']) 
away_data = json_normalize(data= fifa_table['away_team_statistics'])
#combine results     
team_data = home_data.append(away_data)

#plot based on pass accuracy 
team_data[['country', 'attempts_on_goal' , 'pass_accuracy','ball_possession']].sort_values('country', ascending=False)
sns.scatterplot(x="attempts_on_goal", y="ball_possession",  hue = "pass_accuracy",  data=team_data)
plt.legend(loc='right');
```

```python
# Conclusion: Those teams with more attempts on the goal possession have greater ball possession.
#             Greater pass accuracy also increases levels of ball possession.
#             Most attempts on goal are made by those with high pass accuracy.
```

```python
#a

import pandas as pd
bank_data = pd.read_csv("../Data/bank.csv", delimiter = " ", names = ['age', 'sex', 'region', 'income', 'married', 'children', 'car','save_acct', 'current_acct', 'mortgage', 'pep'])
```

```python
numeric_data = bank_data.replace(['NO', 'YES', 'MALE', 'FEMALE'],[0,1,0,1])
numeric_data.head()
```

```python
import numpy as np
numeric_data = numeric_data.values
```

```python
numeric_data[0:5]
```

```python
features = np.zeros((len(numeric_data),4))
numeric_data = np.append(numeric_data,features,1)
j = 2
#one hot decoder  
for i in range(len(numeric_data)):
    if numeric_data[i][j] == 'INNER_CITY':
        numeric_data[i][11:15] = [1,0,0,0]
    if numeric_data[i][j] == 'TOWN':
         numeric_data[i][11:15] = [0,1,0,0]
    if numeric_data[i][j] == 'RURAL':
         numeric_data[i][11:15] = [0,0,1,0]
    if numeric_data[i][j] == 'SUBURBAN':
         numeric_data[i][11:15] = [0,0,0,1]        
#remove redundant column
numeric_data = np.delete(numeric_data, 2, 1)
```

```python
numeric_data[0:5]
```

```python
numeric_data.astype(float)
```

```python
numeric_data.shape
```

```python
from IPython.core.display import Image, display
from IPython.core.display import HTML
from sklearn import tree #For our Decision Tree
from sklearn.tree import DecisionTreeClassifier
from graphviz import Digraph
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
# Exercise 2b asks for different parameters, so we will build a depth-4 tree 
# concurrently with depth-6 to use for the comparison 

dtree4 = DecisionTreeClassifier(max_depth = 4)

dtree6 = DecisionTreeClassifier(max_depth = 6)
bank_labels = numeric_data[:, 9].astype(int)
bank_attrs  = np.delete(numeric_data, 9, 1)
bank_attrs.shape
```

```python
bank_attrs[0:5]
```

```python
bank_labels[0:5]
```

```python
# Splitting the data to train and test data

bank_features_train, bank_features_test, bank_labels_train, bank_labels_test = train_test_split(bank_attrs, bank_labels, test_size=0.33, random_state=42)

dtree4.fit(bank_features_train,bank_labels_train);
dtree6.fit(bank_features_train,bank_labels_train);
```

```python
# Prediction 

predictions_test4 = dtree4.predict(bank_features_test)

predictions_test6 = dtree6.predict(bank_features_test)
predictions_test6[0:10]
```

```python
# Evaluation 

cm4 = metrics.confusion_matrix(bank_labels_test,predictions_test4)
cm6 = metrics.confusion_matrix(bank_labels_test,predictions_test6)
print(cm6)
```

```python
# Precision, Recall, Accuracy

print("Precision (depth 6): ", metrics.precision_score(bank_labels_test,predictions_test6))
print("Recall (depth 6): ", metrics.recall_score(bank_labels_test,predictions_test6))
print("Accuracy (depth 6): ", metrics.accuracy_score(bank_labels_test,predictions_test6))
```

```python
#ROC and AUC on Bank dataset
```

```python
dtree4_prob = dtree4.predict_proba(bank_features_test)

dtree6_prob = dtree6.predict_proba(bank_features_test)
dtree6_prob[0:5]
```

```python
fpr4, tpr4, thresholds4 = metrics.roc_curve(bank_labels_test, dtree4_prob[:,1], pos_label=1, sample_weight=None)
fpr6, tpr6, thresholds6 = metrics.roc_curve(bank_labels_test, dtree6_prob[:,1], pos_label=1, sample_weight=None)
```

```python
# b: plotted depth-6 tree and depth-4 tree (parameter change)


import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
plt.figure(figsize=(5,5))
plt.plot(fpr6,tpr6,linewidth=2.0, color='b', label = "tree-depth6")
plt.plot(fpr4,tpr4,linewidth=2.0, color='r', label = "tree-depth4") # different params
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.xlim([0,1])
plt.ylim([0,1])
ident = [0.0, 1.0]
plt.plot(ident,ident, ls="--", c=".3")
plt.legend();
```
